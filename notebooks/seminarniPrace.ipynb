```python
import re
import csv
file = open("/Users/ondrejbruha/Documents/GitHub/scientificProject/data/dinolist.txt")
lines = file.readlines()
header = lines[0]
header = header.split(",")
feetIndex = header.index('"feet"')
yearIndex = header.index('"year"')
miliIndex = header.index('"million_years"')
clean = []
for l in lines:
    line = l.split(",")
    newLine = []
    for i in line:
        if i == "NA" or i =='"ne"' or i =='"nezname"':
            newLine.append(0)
        elif i == '"detected"':
            newLine.append(True)
        elif line.index(i) == feetIndex or line.index(i) == yearIndex or line.index(i) == miliIndex:
            newLine.append(re.findall("[0-9]+", i))
        else:
            newLine.append(i)
    clean.append(newLine)
with open("/Users/ondrejbruha/Documents/GitHub/scientificProject/data/DinoClean.csv","r+") as file:
    writer = csv.writer(file)
    writer.writerows(clean)
```
